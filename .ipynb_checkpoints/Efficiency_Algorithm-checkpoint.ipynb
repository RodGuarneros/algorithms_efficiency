```python
# Dependencias
import pandas as pd
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import unicodedata
import numpy as np
import math
```

# 1.1. 洧녝(1) 洧녺洧 洧녝(log 洧녵)

```python
# Comparando la O may칰scula considerada como un l칤mite superior en los t칠rminos establecidos previamente:

# Vamos a definir las funciones O(1) y O(log n). Recordemos, de las clases vistas que la funci칩n g(n) nos proporciona una idea de la complegidad del algoritmo, la cual aumenta en funci칩n del tama침o del insumo.

# O(1) o la constante, que no depende del tama침o de la entrada y eso implica que va a ser muy veloz en cualquier momento.


complejidad_O1 = []
def constante(n):
    return 3 # supongamos que la complejidad (g(n)) del algoritmo no cambia y siempre es 3.
    
for i in range(1, 200): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_O1.append(constante(i))

complejidad_O_logn = []
def log_O (n):
    return np.log(n) # supongamos que la complejidad (g(n)) del algoritmo es un logaritmo base 10.
    
for i in range(1, 200): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_O_logn.append(log_O(i))
    
plt.title("Graficando de manera separada O(1) y O(log(n))")
plt.xlabel("Tama침o del insumo (n)")
plt.ylabel("Grado de complejidad g(n)")
      
plt.plot(complejidad_O1, label = "O(1)")
plt.plot(complejidad_O_logn, label = "O(log n)")
plt.legend(loc='best')
plt.grid()

plt.show()

```

```python
plt.title("Comparativo O(1) vs O(log(n))")
plt.xlabel("O(log(n))")
plt.ylabel("O(1)")
      
plt.plot(complejidad_O1, complejidad_O_logn)
#plt.legend(loc='best')
plt.grid()
plt.autoscale()


plt.show()
```

# 1.2. 洧녝(洧녵) 洧녺洧 洧녝(洧녵log洧녵)

```python
# mismos supuestos que los anteriores

complejidad_On = []
def lineal(n):
    return n # supongamos que la complejidad (g(n)) del algoritmo es lineal.
    
for i in range(1, 200): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_On.append(lineal(i))

complejidad_O_nlogn = []
def nlogn (n):
    return n*(np.log(n)) # supongamos que la complejidad es n log n.
    
for i in range(1, 200): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_O_nlogn.append(nlogn(i))
    
plt.title("Recorrido de O(n) y O(nlog(n))")
plt.xlabel("Tama침o del insumo (n)")
plt.ylabel("Grado de complejidad g(n)")
      
plt.plot(complejidad_On, label = 'O(n)')
plt.plot(complejidad_O_nlogn, label = 'O(n log n)')
plt.legend(loc='best')
plt.grid()
plt.savefig("recorrido.png")
plt.show()

```

```python
plt.title("Comparativo O(n) vs O(nlog(n))")
plt.xlabel("O(nlog(n))")
plt.ylabel("O(n)")
      
plt.plot(complejidad_On, complejidad_O_nlogn)
# plt.legend(loc='best')
plt.grid()
plt.autoscale()
# plt.legend()
plt.savefig("comparativo1.png")
plt.show()
```

 # 1.3. 洧녝(洧녵2) 洧녺洧 洧녝(洧녵3)

```python
complejidad_On2 = []
def cuadratica(n):
    return n**2 # supongamos que la complejidad (g(n)) es cuadr치tica.
    
for i in range(1, 200): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_On2.append(cuadratica(i))

complejidad_On3 = []
def cubica (n):
    return n**3 # supongamos que la complejidad es c칰bica.
    
for i in range(1, 200): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_On3.append(cubica(i))
    
plt.title("Recorrido O(n2) y O(n3)")
plt.xlabel("Tama침o del insumo (n)")
plt.ylabel("Grado de complejidad g(n)")
      
plt.plot(complejidad_On2, label = 'O(n2)')
plt.plot(complejidad_On3, label = 'O(n3)')
plt.legend(loc='best')
plt.grid()

plt.show()
```

```python
plt.title("Comparativo O(n2) vs O(n3)")
plt.xlabel("O(n3)")
plt.ylabel("O(n2)")
      
plt.plot(complejidad_On2, complejidad_On3)
# plt.legend(loc='best')
plt.grid()
plt.autoscale()
# plt.legend()
plt.savefig('comparativo2.png')
plt.show()
```

# 1.4. 洧녩洧녵 洧녺洧 洧녵!

```python
complejidad_Oan = []
def lineal2(n):
    return 2**n # supongamos que la complejidad (g(n)) a = 2.
    
for i in range(1, 200): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_Oan.append(lineal2(i))

complejidad_nfact = []
def factorial (n):
    return math.factorial(n) # supongamos que la complejidad es c칰bica.
    
for i in range(1, 200): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_nfact.append(factorial(i))
    
plt.title("Recorrido O(an) y O(n!)")
plt.xlabel("Tama침o del insumo (n)")
plt.ylabel("Grado de complejidad g(n)")
      
plt.plot(complejidad_On2, label = 'O(a^n)')
plt.plot(complejidad_On3, label = 'O(n!)')
plt.legend(loc='best')
plt.grid()

plt.show()
```

```python
plt.title("Comparativo O(an) vs O(n!)")
plt.xlabel("O(n!)")
plt.ylabel("O(an)")
      
plt.plot(complejidad_On2, complejidad_On3)
# plt.legend(loc='best')
plt.grid()
plt.autoscale()
# plt.legend()

plt.show()
```

# 1.5. 洧녵! 洧녺洧 洧녵洧녵

```python
complejidad_nfact = []
def factorial (n):
    return math.factorial(n) # supongamos que la complejidad es c칰bica.
    
for i in range(1, 10): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_nfact.append(factorial(i))
    
complejidad_nn = []
def producto(n):
    return n*n # supongamos que la complejidad es producto de si mismo.
    
for i in range(1, 10): # supongamos que el tama침o de los isumos pueden tomar valores de 1 a 200.
    complejidad_nn.append(producto(i))
    
plt.title("Recorrido O(n!) y O(nn)")
plt.xlabel("Tama침o del insumo (n)")
plt.ylabel("Grado de complejidad g(n)")
      
plt.plot(complejidad_nfact, label = 'O(n!)')
plt.plot(complejidad_nn, label = 'O(n*n)')
plt.legend(loc='best')
plt.grid()

plt.show()
```

```python
plt.title("Comparativo O(n!) vs O(nn)")
plt.xlabel("O(an)")
plt.ylabel("O(n!)")
      
plt.plot(complejidad_nfact, complejidad_nn)
# plt.legend(loc='best')
plt.grid()
plt.autoscale()
# plt.legend()

plt.show()
```

# 2. Haga una tabla donde simule tiempos de ejecuci칩n suponiendo que cada operaci칩n tiene un costo de 1 micro-segundo:

#### - Suponga que cada uno de los ordenes de crecimiento anteriores es una expresi칩n g(n) que describe el costo de un algoritmo teniendo en cuenta el tama침o de la entrada del algoritmo 洧녵. Cada operaci칩n expresada en el eje de las y, equivale a un micro-segundo, de modo tal que si dividimos el g(n)/1000000 se tendr칤a el costo expresado en segundos en que incurre el algoritmo para procesar el insumo n.

```python
N = [100, 1000, 10000, 100000]

# O(1) = []
# log(n) = []
# n = []
# nlogn = []
# n^2 = []
# n^3 = []
# a^n = []
# factorial = []

A = []
E = []
I = []
O = []
U = []
R = []
S = ['4X10^6 Cen','1X10^166 Cen','1X10^1666 Cen','1X10^16666 Cen']
T = ['3X10^144 Cen','1X10^2554 Cen','1X10^25545 Cen','1X10^2554554 Cen']

# Calculamos el costo por segundo, minuto, semana

for n in N:
    A.append(constante(n)) # micro-segundos.
    E.append(log_O(n)) # micro-segundos.
    I.append(lineal(n)/1000000) # Es posible expresar el algoritmo lineal en segundos 1e+6 micro-segundos por segundo.
    O.append(nlogn(n)/1000000) # Es posible expresar el algoritmo lineal en segundos 1e+6 micro-segundos por segundo.
    U.append(cuadratica(n)/60000000) # Demasiado grande, es posible expresar el algoritmo lineal en minutos 6e+7 micro-segundos por minuto.
    R.append(cubica(n)/604800000000) # Demasiado grandes para segundos se transforman en semanas, tomando en cuenta que hay 6.048e+11 micro-segundos por semana 
    # S.append(lineal2(n)/(604800000000) # no me fue posible hacer las operaciones, pero encontr칠 las equivalencias en Internet.
    # T.append(factorial(n)/1000000) # no me fue posible hacer las operaciones, pero encontr칠 las equivalencias en Internet.
```

```python
Lists = [A, E, I, O, U, R, S, T]

df = pd.DataFrame(Lists)

df.columns = ['100','1000','10000','100000']

df = df.rename(index={0: 'O(1) micro-segundos', 1: 'O(log(n)) micro-segundos', 2: 'O(n) segundos', 3: 'O(n log n) segundos', 4: 'O(n^2) minutos', 5: 'O(n^3) semanas', 6: 'O(a^n) Cententillones', 7: 'O(n!) Centillones'})

df
```
